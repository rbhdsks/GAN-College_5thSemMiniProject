```python
import os
import shutil

# Path to your images directory
source_dir = '/content/drive/MyDrive/ToDayGAN/results/fs/test_15/images'

# Paths for the new directories to store segregated images
real_dir = '/content/drive/MyDrive/ToDayGAN/results/fs/test_15/real_images'
fake_dir = '/content/drive/MyDrive/ToDayGAN/results/fs/test_15/fake_images'

# Create directories if they don't exist
os.makedirs(real_dir, exist_ok=True)
os.makedirs(fake_dir, exist_ok=True)

# Loop through all files in the source directory
for filename in os.listdir(source_dir):
    if 'real' in filename:
        shutil.move(os.path.join(source_dir, filename), real_dir)
    elif 'fake' in filename:
        shutil.move(os.path.join(source_dir, filename), fake_dir)

print('Files have been segregated into real and fake directories.')
```
